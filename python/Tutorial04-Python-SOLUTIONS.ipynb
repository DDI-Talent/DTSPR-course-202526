Within the `Hbr` variable, "S92000003" is the country code for Scotland. The `Sex` variable we do not need for this task, but it too includes an aggregate level "All". Filter the data such that only these aggregate level are included for these 2 variables. Filter the data accordingly and save this processed dataset into an object called `stroke`.

#### Task 12 solution 

```python
stroke = stroke_raw.loc[(stroke_raw["Hbr"] == "S92000003") & 
                       (stroke_raw["Sex"] == "All")].loc[:,["FinancialYear", "AdmissionType", 
                                                            "AgeGroup", "Diagnosis",
                                                            "NumberOfDischarges"]]

print(stroke)
```

```python
## you could also use isin instead of == to filter HBR and Sex
stroke_raw.loc[(stroke_raw["Hbr"].isin(["S92000003"])) & 
              stroke_raw["Sex"].isin(["All"])].loc[:,["FinancialYear", "AdmissionType",
                                                      "AgeGroup", "Diagnosis",
                                                      "NumberOfDischarges"]]
```

Notice our index still shows 43199 indexed as the last value, though we only have 960 rows in the `stroke` data frame. If we specify the argument `drop = True` the index column will be dropped, but by default the original index is added as a column to the dataframe. ## Task 15 

`AgeGroup` is a bit messy. It should be ordered and includes 2 aggregate categories. Remove the aggregate categories and order the remaining categories.

#### Task 15 Solution

### Task 16 

Create a summary table with the average number of discharges with a stroke diagnosis by age group for all admissions in the financial years 2021/22 and 2022/23.

<details><summary style='color:darkblue'>HINT: Breaking down the task. CLICK HERE TO SEE</summary>

Further filtering of the data is needed for this task. THEN grouping the data in order to compute our values of interest


```python
## your answer 

```

#### Task 16 solution

```python
stroke.loc[(stroke["AdmissionType"] == "All") & 
           (stroke["Diagnosis"]== "stroke") & 
           (stroke["FinancialYear"].isin(["2021/22", "2022/23"]))].groupby(["FinancialYear", "AgeGroup"], observed = True)["NumberOfDischarges"].mean().dropna()

# change the order of the variables in the groupby list to see how the output changes 
```

*📝 Note*: remove `observed = True` from the `.groupby()` chain and read the warning. This highlights the importance of not only reading warnings, but in understanding data types! Have a quick read through [the documentation for groupby here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html), focusing on the parameters section and what is written for the `observed` parameter. 


... the default behavior will be changing to `True` in a future version of `pandas`, hence the warning alerting us to this change in behavior. Regarding data types, this parameter is important when dealing with Cagtegorical data, which is the case here and you are likely to do in many cases. "This only applies if any of the groupers are Categoricals. If True: only show observed values for categorical groupers. If False: show all values for categorical groupers."

Another approach, if you wanted to break the code down is to use *intermediate objects*. Demonstrated below:

```python
# create an object with the data filtered 
stroke_filter = stroke.loc[(stroke["AdmissionType"] == "All") & 
           (stroke["Diagnosis"]== "stroke") & 
           (stroke["FinancialYear"].isin(["2021/22", "2022/23"]))]

# then using that filtered object we can apply the group by 
stroke_groupby = stroke_filter.groupby(["FinancialYear", "AgeGroup"], observed = True)

# then finally from the filtered group by object, get the mean for our solution
stroke_final = stroke_groupby["NumberOfDischarges"].mean().dropna()

print(stroke_final)
```

### Bonus - consistent column names  

Sometimes the formatting of column names in a dataframe is not consistent, which can cause errors and issues in your code. For example, in the January 2025 update to the data, the PHS team have changed the title case of the health board region column, which used to be `HBR` and is now `Hbr`. This will cause bugs in any code written previously using `HBR`. This is one example of why, in the process of reading in data, it can be a good idea to reformat column names for consistency (e.g., to be all lowercase). **Can you think of how we might programmatically make all column names lowercase?** 

I can think of at least 3 approaches, share on the discussion boards if you come up with others! 

```python
# 1st make a copy of the df - we will make 3 to try out the 3 approaches
stroke_copy1 = stroke_raw.copy()

stroke_copy2 = stroke_raw.copy()

stroke_copy3 = stroke_raw.copy() 
```

```python
## Approach 1 - use str.lower on the columns attribute of the data frame 
stroke_copy1.columns = stroke_copy1.columns.str.lower()

print(stroke_copy1)
```

```python
## Approach 2 - use the pd.DataFrame.rename function and str.lower 
### this can be useful as it returns a DataFrame and allows in-place modification 
# Documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html 

stroke_copy2 = stroke_copy2.rename(str.lower, axis = 'columns')

print(stroke_copy2)
```

```python
## Approach 3 (more advanced) - use map and lambda functions to iterate the desired transformation 

stroke_copy3.columns = stroke_copy3.columns.map(lambda colnames: colnames.lower()) 

print(stroke_copy3)
```

The 3rd approach is somewhat over the top in this scenario, but is an extremely useful superpower in the form of lambda functions (which are small **anonymous functions**: a temporary function that is not bound to a specific name or identifier). The power of lambda are best seen as above when you use them as an anonymous function inside another function.

The summary here is that (once again) **there are many ways to do the same thing - pick the one that makes the most sense to you and you can explain/understand the best.**
